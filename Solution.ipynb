```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("used_cars_uk.csv")
data.head(10)
```

```python
data.info()
data.describe()
```

```python
def plot_scatter(data, x, y, hue=None, title="Scatter Plot"):
    """
    Function to create a scatter plot.
    
    :param data: DataFrame containing the data
    :param x: String representing the column for the x-axis
    :param y: String representing the column for the y-axis
    :param hue: String representing the column for color encoding
    :param title: String representing the title of the plot
    """
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x=x, y=y, hue=hue)
    plt.title(title)
    plt.show()

# Usage
plot_scatter(data, 'Mileage(miles)', 'Price', hue='Fuel type')
```

```python
def plot_histogram(data, column, title="Histogram"):
    """
    Function to create a histogram.
    
    :param data: DataFrame containing the data
    :param column: String representing the column for the histogram
    :param title: String representing the title of the plot
    """
    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], kde=False)
    plt.title(title)
    plt.show()

# Usage
plot_histogram(data, 'Price')
```
